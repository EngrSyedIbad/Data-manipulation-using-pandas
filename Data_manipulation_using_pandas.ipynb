# **SLICING & INDEXING**

```python
a=['spam','egg','bacon','tomato','ham','lobster']
#0 - n-1 positive indexing
#-1 - n negative indexing
```

```python
a[0]
```

```python
len(a)
```

```python
a[len(a)-1]
```

```python
a[-6] #negative indexing
```

**SLICING**

```python
a=['spam','egg','bacon','tomato','ham','lobster']
```

```python
a[2:5] #wont give last indexed variable
```

```python
a[-5:-2] #last se pehel
```

```python
a[:6] #assumes first index
```

```python
a[0:6]
```

```python
a[2:] # assumes end
```

```python
a[:]
```

**stride**

```python
a #stride
```

```python
a[0:10:2]  #slicing usimg strides skipping 1 number
```

```python
a[6:0:-2] #negative stride reverse
```

```python
a[::-1] #reserving
```

# **SELECTION loc and iloc**

```python
import pandas as pd
df=pd.read_csv('titanic.csv')
df.head(2)
```

```python
df.Pclass
```

```python
df['Pclass'][0]
```

```python
df.iloc[0] #rows
```

```python
df.iloc[0:3,4]
```

```python
df.iloc[[0,2],4]
```

```python
df.iloc[-5:] #reverse order descending
```

**Label Based Selection**

```python
df.loc[0,'Name'] #loc is inclusive 0 to 10 means 0 to 10 whereas iloc is exclusive 0 to 10 means 0 to 9
```

```python
df.loc[:,['Name','Sex','Fare']]
```

**Conditional Selection**

```python
df.Survived==1
```

```python
df.loc[df.Survived ==1]
```

```python
df.loc[(df.Survived ==1)&(df.Sex=='male')]
```

```python
df.loc[(df.Survived ==1)|(df.Sex=='male')]
```

```python
df.loc[df.Embarked.isin(['S','Q'])]
```

```python
df.loc[df.Fare.between(10,20)]
```

```python
df.loc[df.Fare.notnull()]
```

# **RESHAPING DATA**

```python
import numpy as np
df.groupby('Pclass').Fare.mean()
```

```python
df.groupby(['Sex','Pclass']).Fare.mean()
```

```python
df.groupby(['Sex','Pclass']).Age.mean()
```

```python
df.groupby('Pclass').Fare.agg([np.min,np.max,np.mean,np.median])
```

```python
df.groupby('Pclass').agg({'Fare':['min','max','mean','median'],'Age':['min','max','mean','median']})
```
